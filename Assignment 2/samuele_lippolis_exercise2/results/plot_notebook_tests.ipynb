```python
import matplotlib.pyplot as plt
import pandas as pd
from io import StringIO
```

```python
# Your data
data = """Matrix Size,Elapsed Time (s),GFLOPS
2000,.16719,957.27622
4000,.89547,1429.41012
6000,.28527,1514.33086
8000,.64725,1582.06145
10000,1.28917,1551.39805
12000,2.23790,1544.64343
14000,3.47409,1579.98468
16000,5.23930,1563.60688
18000,7.62151,1531.03224
20000,10.37138,1542.71874"""

# Create a DataFrame from the data
df = pd.read_csv(StringIO(data))

# Extract the columns
matrix_size = df['Matrix Size']
elapsed_time = df['Elapsed Time (s)']
gflops = df['GFLOPS']

# Create the first plot (matrix size vs. time)
plt.figure(figsize=(8, 4))
plt.plot(matrix_size, elapsed_time, marker='o', linestyle='-')
plt.xlabel('Matrix Size')
plt.ylabel('Elapsed Time (s)')
plt.title('Matrix Size vs. Elapsed Time')
plt.grid(True)
plt.show()

# Create the second plot (matrix size vs. GFLOPS)
plt.figure(figsize=(8, 4))
plt.plot(matrix_size, gflops, marker='o', linestyle='-')
plt.xlabel('Matrix Size')
plt.ylabel('GFLOPS')
plt.title('Matrix Size vs. GFLOPS')
plt.grid(True)
plt.show()
```

```python
# Your data
data = """Matrix Size,Iteration,Elapsed Time (s),GFLOPS
2000,1,0.17264148,926.776114
2000,2,0.16380919,976.746176
2000,3,0.16523637,968.309822
2000,4,0.16670968,959.752307
2000,5,0.16757494,954.796702
4000,1,0.89321909,1433.019082
4000,2,0.89393014,1431.879229
4000,3,0.89828961,1424.930207
4000,4,0.89713370,1426.766155
4000,5,0.89481958,1430.455959
6000,1,0.285424756,1513.533745
"""

# Create a DataFrame from the data
df = pd.read_csv(StringIO(data))

# Compute the sample mean and standard deviation for time and GFLOPS
time_mean = df.groupby('Matrix Size')['Elapsed Time (s)'].mean()
time_std = df.groupby('Matrix Size')['Elapsed Time (s)'].std()

gflops_mean = df.groupby('Matrix Size')['GFLOPS'].mean()
gflops_std = df.groupby('Matrix Size')['GFLOPS'].std()

# Create the first plot (matrix size vs. sample mean of time with shaded standard deviation)
plt.figure(figsize=(8, 4))
plt.plot(time_mean.index, time_mean, marker='o', linestyle='-', label='Mean')
plt.fill_between(time_mean.index, time_mean - time_std, time_mean + time_std, alpha=0.2, label='Standard Deviation')
plt.xlabel('Matrix Size')
plt.ylabel('Sample Mean of Elapsed Time (s)')
plt.title('Matrix Size vs. Sample Mean of Elapsed Time')
plt.grid(True)
plt.legend()
plt.show()

# Create the second plot (matrix size vs. sample mean of GFLOPS with shaded standard deviation)
plt.figure(figsize=(8, 4))
plt.plot(gflops_mean.index, gflops_mean, marker='o', linestyle='-', label='Mean')
plt.fill_between(gflops_mean.index, gflops_mean - gflops_std, gflops_mean + gflops_std, alpha=0.2, label='Standard Deviation')
plt.xlabel('Matrix Size')
plt.ylabel('Sample Mean of GFLOPS')
plt.title('Matrix Size vs. Sample Mean of GFLOPS')
plt.grid(True)
plt.legend()
plt.show()
```
